```python
import torch
```

```python
# Define the ReLU activation function
def relu(x):
    return x * (x > 0).float()
```

```python
x = torch.tensor([-1.0, 0.0, 1.0, 2.0])
print("Input: ", x)
print("Output: ", relu(x))
```
